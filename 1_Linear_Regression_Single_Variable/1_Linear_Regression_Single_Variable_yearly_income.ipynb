```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
X = df[['year']]
```

```python
y = df['per capita income (US$)']
```

```python
plt.scatter(X, y, color='r', marker='+')
plt.xlabel('year')
plt.ylabel('income (US$)')
```

**Train the model using X and y:**

```python
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

**Predict the income for the year of 2020 using trained model:**

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y, label='Actual Data', color='r', marker='*')
plt.plot(X, reg.predict(X), label='Predicted value', color='g')
plt.xlabel('year')
plt.ylabel('income (US$)')
plt.legend()
plt.show()
```
