# Logistic Regression (Binary Classification)

**Importing Libraries:**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Loading and Inspecting Data:**

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

**Visualizing Data:**

```python
plt.scatter(df.age, df.bought_insurance, marker='+', color='r')
plt.xlabel('Age')
plt.ylabel('Bought Insuranc')
plt.title('Whether a customer will buy insurance based on their age')
```

**Checking Dataset Shape (it gives the number of rows and columns in the dataset. This helps us understand the size of the dataset):**

```python
df.shape
```

**Splitting Data into Training and Testing Sets (The model learns from training data and is tested on unseen test data):**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9)
```

**Checking Test Data:**

```python
X_test
```

**Training the Model (`LogisticRegression()` creates a model that predicts probabilities for classification tasks):**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

**This predicts whether each person in `X_test `(test data) will buy insurance (`1` or `0`):**

```python
model.predict(X_test)
```

**Checking Model Accuracy:**

```python
model.score(X_test, y_test)
```

**Checking Prediction Probabilities(Instead of just 0 or 1, this gives probabilities for each class. For example: [0.2, 0.8]  20% chance of 0, 80% chance of 1.):**

```python
model.predict_proba(X_test)
```

**Predicting for a Specific Age:**

```python
model.predict([[25]])
```
