# **KFold Cross Validation Exercise**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

import pandas as pd

from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

### Training the Models & Cross-Validation

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=5)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_results_l = cross_val_score(LogisticRegression(), X, y, cv=folds)
```

```python
print("Cross Validation Results for Logistic Regression (Accuracy):")
for i, result in enumerate(cross_val_results_l):
  print(f"Fold {i+1}: {result * 100 :.2f}%")
print(f'Mean Accuracy: {cross_val_results_l.mean() * 100 :.2f}%')
```

```python
cross_val_results_svm = cross_val_score(SVC(C=5), X, y, cv=folds)
```

```python
print("Cross Validation Results for Support Vector Machine (Accuracy):")
for i, result in enumerate(cross_val_results_svm):
  print(f"Fold {i+1}: {result * 100 :.2f}%")
print(f"Mean Accuracy: {cross_val_results_svm.mean() * 100 :.2f}%")
```

```python
cross_val_results_rf = cross_val_score(RandomForestClassifier(n_estimators=40), X, y, cv=folds)
```

```python
print("Cross Validation Results for Random Forest Classifier (Accuracy):")
for i, result in enumerate(cross_val_results_rf):
  print(f"Fold {i+1}: {result * 100 :.2f}%")
print(f"Mean Accuracy: {cross_val_results_rf.mean() * 100 :.2f}%")
```
