```python
import pandas as pd

df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.shape
```

```python
df.price.describe()
```

```python
min_threshold = df['price'].quantile(0.01)
min_threshold
```

```python
df[df['price'] < min_threshold]
```

```python
max_threshold = df.price.quantile(0.99)
max_threshold
```

```python
df[df['price'] > max_threshold]
```

```python
main_df = df[(df.price > min_threshold) & (df.price < max_threshold)]
```
