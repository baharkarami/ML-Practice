```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import math

def predict_using_sklean():
  df = pd.read_csv('test_scores.csv')
  reg = linear_model.LinearRegression()
  reg.fit(df[['math']], df.cs)
  return reg.coef_, reg.intercept_


def predict_using_gradient_decsent(x,y):
  m_curr = b_curr = 0
  iterations = 1000000
  learning_rate = 0.0002
  n = len(x)
  previous_cost = 0

  for i in range(iterations):
    y_predicted = m_curr * x + b_curr
    cost = (1/n) * sum([value**2 for value in (y - y_predicted)])
    m_derivation = -(2/n) * sum(x * (y - y_predicted))
    b_derivation = -(2/n) * sum(y - y_predicted)
    m_curr = m_curr - learning_rate * m_derivation
    b_curr = b_curr - learning_rate * b_derivation

    if math.isclose(cost, previous_cost, rel_tol=1e-20):
        break
    previous_cost = cost

    print("m {}, b {}, cost {}, iteration {} \n".format(m_curr, b_curr, cost, i))

  return m_curr, b_curr

if __name__ == "__main__":
  df = pd.read_csv('test_scores.csv')
  x = np.array(df.math)
  y = np.array(df.cs)

  m, b = predict_using_gradient_decsent(x,y)
  print("using gradient decsent: coef = {}, intercept = {}".format(m, b))

  m_sklearn, b_sklearn = predict_using_sklean()
  print("using sklearn: coef = {}, intercept = {}".format(m_sklearn, b_sklearn))

```
