# **Random Forest Exercise**

**Loading Iris Dataset:**

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

**Listing All Available Attributes of `iris`:**

```python
dir(iris)
```

**Creating a DataFrame out of `iris`:**



```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

**Adding the `target` Column to `df`, Which Represents the Kind of Iris (0=setosa, 1=versicolor, 2=virginica):**

```python
df['target'] = iris.target
df.head()
```

**Splitting Data into Features (`X`) and Labels (`y`):**

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

**Splitting Data into Training and Testing Sets:**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**Training a Random Forest Classifier (with 15 decision trees (`n_estimators=15)`):**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=15)
model.fit(X_train, y_train)
```

**Evaluate the Model**

```python
model.score(X_test, y_test)
```

**Computing Confusion Matrix:**

```python
y_predicted = model.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

**Visualizing Confusion Matrix:**

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
