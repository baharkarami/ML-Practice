# Logistic Regression (Multiclass Classification)

**Importing Necessary Libraries:**

```python
import matplotlib.pyplot as plt
from sklearn import datasets
```

**Loading the Digits Dataset (This loads the digits dataset, which contains 1,797 grayscale images (8×8 pixels) of handwritten digits (0-9). Each image is represented as a 64-element array (flattened from 8×8)):**

```python
digits = datasets.load_digits()
```

**Checking Dataset Contents:**

```python
dir(digits)
```

**This retrieves the first image's pixel values (a flattened array of 64 numbers):**

```python
digits.data[0]
```

**Visualizing the First Image:**

```python
plt.gray()
plt.matshow(digits.images[0])
```

**Displaying the First 5 Images:**

```python
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
```

**Checking First 5 Target Labels:**

```python
digits.target[0:5]
```

**Splitting Data into Training and Testing Sets:**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

**Checking Training and Testing Sizes:**

```python
len(X_train)
```

```python
len(X_test)
```

**Training the Logistic Regression Model:**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

**Evaluating Model Accuracy:**

```python
model.score(X_test, y_test)
```

**Displaying the 67th image:**

```python
plt.matshow(digits.images[67])
```

**Retrieving the 67th image true label:**

```python
digits.target[67]
```

**Uses `model.predict()` to check if the model correctly predicts the digit:**

```python
model.predict([digits.data[67]])
```

**Predicting Multiple Digits**

```python
model.predict(digits.data[0:5])
```

**Creating a Confusion Matrix (The confusion matrix helps analyze the model’s mistakes by comparing actual vs. predicted labels):**

```python
y_predicted = model.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

**Visualizing the Confusion Matrix:**

```python
import seaborn as sn
plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

for example colored 35 means 35 times the truth was 0 and my model predict it as 0.
