```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('carprices.csv')
df
```

**Car Mileage Vs Sell Price ($)**

```python
plt.scatter(df['Mileage'], df['Sell Price($)'])
plt.xlabel('Mileage')
plt.ylabel('Sell Price')
```

**Car Age Vs Sell Price ($)**

```python
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
plt.xlabel('Age')
plt.ylabel('Sell Price')
```

**The scatter plots show a clear linear relationship between the dependent variable (Sell Price) and independent variables (Car Age & Mileage), making linear regression a good choice.**

```python
X = df[['Mileage', 'Age(yrs)']]
X
```

```python
y = df['Sell Price($)']
y
```

**We'll split the data into:**

  * **Training Set: Used to train the model.**
  * **Testing Set: Used to evaluate its predictions.**

**Testing on the same data we trained on would be misleading, like giving students the exact same questions from class in an examit wouldnt truly measure performance.**

**Imports the function to split data into training and testing sets:**

```python
from sklearn.model_selection import train_test_split
```

**Splits dataset (`X, y`) into training (70%) and testing (30%) sets:**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

**Imports the LinearRegression model. Creates an instance of the linear regression model:**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

**Trains the model using the training data:**

```python
model.fit(X_train, y_train)
```

**Uses the trained model to predict `y` values for the test set:**

```python
model.predict(X_test)
```

**Displays the actual y values for the test set:**

```python
y_test
```

**Returns the model's accuracy (R score) on the test data:**

```python
model.score(X_test, y_test)
```
